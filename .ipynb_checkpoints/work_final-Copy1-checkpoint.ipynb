# Aeropuertos y puertos

```python
import geopandas as gpd
import os
from  fiona import listlayers
import pandas as pd
airports= os.path.join("maps","hotosm_niger_airports_gpkg","hotosm_niger_airports.gpkg")
otros_puntos=os.path.join("maps","hotosm_niger_sea_ports_gpkg","hotosm_niger_sea_ports.gpkg")
localidades=os.path.join("maps","kontur_topology_boundaries_NE_20230628.gpkg","kontur_topology_boundaries_NE_20230628.gpkg")
# estas son las subcapas que tiene "airports" y "puertos"
airports_1=gpd.read_file(airports,layer='Airports')  #punto
puertos=gpd.read_file(otros_puntos,layer='Sea Ports')   #punto
div_admin=gpd.read_file(localidades,layer='kontur_topology_boundaries_NE_20230628')   # polígono
#puertos.plot()
#airports_1.plot()
airports_1=airports_1.to_crs(2931)
puertos=puertos.to_crs(2931) 
# MATRIZ DE DISTANCIAS
distancia_matriz_port_airp=puertos.set_index('osm_id').geometry.apply\
(lambda g: airports_1.set_index('osm_id').geometry.distance(g)/1000).\
sort_index(axis=0).sort_index(axis=1)
# Distancia de una de las FILAS de la MATRIZ DE DISTANCIAS (en km)
airports_1.iloc[0].geometry.distance(puertos.iloc[0].geometry)/1000
# Distancia mínima entre dos puntos (encima del país)
salidas=pd.DataFrame()
salidas['min']=distancia_matriz_port_airp.min(axis=1)
salidas.head()
# imágenes
```

# Instalaciones educativas y Río Niger

```python
# Distancia entre puntos y ríos
countries=gpd.read_file(os.path.join("maps","World_Countries","World_Countries.shp"))
rivers=gpd.read_file(os.path.join("maps","World_Hydrography","World_Hydrography.shp"))
instalac_educ=gpd.read_file(os.path.join("maps","instalac_educat","hotosm_niger_education_facilities_points.shp"))
niger=countries[countries.COUNTRY=='Niger']
rivers_n=gpd.clip(gdf=rivers,
                  mask=niger) # río de Niger
rivers_n.head()
#instalac_educ.head(10)
```

```python
# distancias entre las inst. educat al río Niger
rivers_n[rivers_n.NAME.str.contains('Niger')].iloc[0].geometry.distance(instalac_educ.set_index('name').geometry)/1000
```

```python
#Eliminación de duplicados
rivers_n=rivers_n.drop_duplicates(subset="NAME")
instalac_educ=instalac_educ.drop_duplicates(subset="name")
# MATRIZ DE DISTANCIAS de todos los ríos con todos las instalaciones educativas
dist_matr_riv_inst_ed=rivers_n.set_index('NAME').geometry.apply\
(lambda g: instalac_educ.set_index('name').geometry.distance(g)/1000).\
sort_index(axis=0).sort_index(axis=1)
dist_matr_riv_inst_ed
```

```python
# distancia con todos los demás instituciones educativas respecto al Río Niger
dist_matr_riv_inst_ed.loc['Niger'].sort_values()
```

```python
mins=dist_matr_riv_inst_ed.idxmin(axis="columns")
```

```python
base=rivers_n.explore()
# Instalación educativa más cercana al río
instalac_educ[instalac_educ.name.isin(mins)].explore(m=base,color='red',marker_kwds=dict(radius=15))
# Instalaciones educativas más lejanas al río
instalac_educ[~instalac_educ.name.isin(mins)].explore(m=base,color='blue',marker_kwds=dict(radius=5))
```

# Lugares administrados e instituciones educativas

```python
# Polígonos y puntos 
dist_matriz_div_inst=div_admin.set_index('admin_level').geometry.apply\
(lambda g: instalac_educ.set_index('name').geometry.distance(g)/1000).\
sort_index(axis=0).sort_index(axis=1)

dist_matriz_div_inst
```

```python
mins=dist_matriz_div_inst.idxmin(axis="columns") 
```

```python
div_adm_M=div_admin[div_admin['admin_level']==2]
div_adm_marina=div_adm_M.convex_hull
base=div_adm_marina.explore()
instalac_educ[instalac_educ.name.isin(mins)].explore(m=base,color='red',marker_kwds=dict(radius=10))
instalac_educ[~instalac_educ.name.isin(mins)].explore(m=base,color='pink',marker_kwds=dict(radius=5))
```

```python
# menor distancia en metros
dist_matr_riv_inst_ed.loc['Niger'].min()
```

# Datos de objetos espaciales 

```python
# Ejercicio 5

#div_adm_M.info()
import geopandas as gpd

DataNiger=os.path.join("maps","ner_adm02_feb2018.zip")

data_Niger=gpd.read_file(DataNiger)

data_Niger.head(5)
```

```python
nigerData=os.path.join("data","ngr_20230930-refugee-asylum-seeker-others-par-commune.xlsx")
datad=pd.read_excel(nigerData,
                     dtype={'Ubigeo': object})
datad.head(5)
```

```python
# MERGE
data_Niger.merge(datad, left_on='adm_01', right_on='coa_admin1').shape
```

```python
capitalizeColumns=lambda x: x.str.upper().str.strip()
datad[['coa_admin2','coa_admin1']]=datad[['coa_admin2','coa_admin1']].apply(capitalizeColumns)
data_Niger[['adm_02','adm_01']]=data_Niger[['adm_02','adm_01']].apply(capitalizeColumns)
```

```python
datad.coa_admin1.duplicated().sum(),data_Niger.adm_01.duplicated().sum()
```

```python
datad['Admin']=["+".join(pd) for pd in zip (datad.coa_admin2,datad.coa_admin1)]
data_Niger['Admin']=["+".join(pd) for pd in zip (data_Niger.adm_02,data_Niger.adm_01)]
```

```python
datad['Admin']=datad.Admin.str.replace("\-|\_|\s+"," ",regex=True)
data_Niger['Admin']=data_Niger.Admin.str.replace("\-|\_|\s+"," ",regex=True)
```

```python
# Las no coincidencias
```

```python
nomatch_df=set(datad.Admin)- set(data_Niger.Admin)
nomatch_gdf=set(data_Niger.Admin)-set(datad.Admin) 
```

```python
len(nomatch_df), len(nomatch_gdf)
```

```python
# fuzzy merging
```

```python
from thefuzz import process
[(dis,process.extractOne(dis,nomatch_gdf)) for dis in sorted(nomatch_df)]
```

```python
{dis:process.extractOne(dis,nomatch_gdf)[0] for dis in sorted(nomatch_df)}
changesDis_df={dis:process.extractOne(dis,nomatch_gdf)[0] for dis in sorted(nomatch_df)}
```

```python
datad.Admin.replace(changesDis_df,inplace=True)
```

```python
datadis_Niger=data_Niger.merge(datad, on='Admin')
```

```python
datadis_Niger.info()
```

```python
bye=['adm_01', 'rowcacode1', 'rowcacode2','Admin','ISO2','ISO3','f_NA','m_NA','source','aggregation_type','accommodation_type']
datadis_Niger.drop(columns=bye,inplace=True)
```

```python
datadis_Niger['f_60'] = datadis_Niger['f_60'].fillna(0)
```

# Lugar vecino

```python
# ejercicio 7
from libpysal.weights import Queen, Rook, KNN
w_rook = Rook.from_dataframe(datadis_Niger,use_index=False)
w_queen = Queen.from_dataframe(datadis_Niger,use_index=False)
w_knn = KNN.from_dataframe(datadis_Niger, k=4)
```

```python
datadis_Niger.iloc[w_rook.neighbors[1] ,].plot(facecolor="yellow")
```

```python
#  ROOK
base=datadis_Niger[datadis_Niger.adm_02=="ABALA"].plot()
datadis_Niger.iloc[w_rook.neighbors[1] ,].plot(ax=base,facecolor="yellow",edgecolor='k')
datadis_Niger.head(1).plot(ax=base,facecolor="red")
```

```python
#  QUEEN
base=datadis_Niger[datadis_Niger.adm_02=="ABALA"].plot()
datadis_Niger.iloc[w_queen.neighbors[1] ,].plot(ax=base,facecolor="yellow",edgecolor='k')
datadis_Niger.head(1).plot(ax=base,facecolor="red")
```

```python
#  KNN
base=datadis_Niger[datadis_Niger.adm_02=="ABALA"].plot()
datadis_Niger.iloc[w_knn.neighbors[1] ,].plot(ax=base,facecolor="yellow",edgecolor='k')
datadis_Niger.head(1).plot(ax=base,facecolor="red")
```
